```python
# importing dataset
import pandas as pd
df = pd.read_csv('/content/Video_Games_Sales_as_at_22_Dec_2016.csv')
dx=df.dropna()
dx
```

```python
#Separate the independent variable(s) (X) and dependent variable (y) from the dataset.
X = df['Critic_Score'].values.reshape(-1,1)
y = df['Global_Sales'].values.reshape(-1,1)
dx


```

```python
#Create an instance of the LinearRegression class from scikit-learn and fit it to the data
from sklearn.linear_model import LinearRegression as lr
model = lr()
```

```python

# the trained model to make predictions on new data.
lr = LinearRegression()
lr.fit(X, y)
```

```python
#the trained model to make predictions on new data.y_pred = lr.predict(X)
y_pred
```

```python
#Finally, evaluate the performance of the model by calculating metrics such as the R-squared value
from sklearn.metrics import r2_score

r2 = r2_score(y, y_pred)


print(f"R-squared: {r2}")

```
